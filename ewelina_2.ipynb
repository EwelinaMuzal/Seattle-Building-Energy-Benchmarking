```python
import pandas as pd
```

```python
df2016 = pd.read_csv("2016-building-energy-benchmarking.csv")
df2017 = pd.read_csv("2017_Building_Energy_Benchmarking.csv")
```

```python
df2017.head()
```

```python
df2016.head()
```

```python
#columns to drop: City, State, PropertyGFABuilding(s), TaxParcelIdentificationNumber + 2016 Comments
drop_2016 = ['City', 'State', 'PropertyGFABuilding(s)', 'TaxParcelIdentificationNumber', 'Comments']
drop_2017 = ['City', 'State', 'PropertyGFABuilding(s)', 'TaxParcelIdentificationNumber']
```

```python
df2016.drop(drop_2016, axis=1, inplace=True)
```

```python
df2016.head()
```

```python
df2017.drop(drop_2017, axis=1, inplace=True)
```

```python
df2017.head()
```

```python
df2016.columns
```

```python
columns_2016 = df2016.columns.tolist()
```

```python
df2017 = df2017[columns_2016]
```

```python
df_16_17 = pd.concat([df2016, df2017], axis=0)
```

```python
df_16_17.info()
```

```python
#SiteEUI(kBtu/sf) : NaturalGas(kBtu)
to_zero = df_16_17.loc[:, 'SiteEUI(kBtu/sf)':'NaturalGas(kBtu)'].columns.tolist()
```

```python
#df_2016_17[to_zero] = df['DataFrame Column'].fillna(0)
```

```python
to_zero
```

```python
df_16_17[to_zero] = df_16_17[to_zero].fillna(0)
```

```python
df_16_17 = df_16_17[df_16_17['Electricity(kWh)'] != 0]
df_16_17 = df_16_17[df_16_17['SiteEnergyUse(kBtu)'] != 0]
```

```python
df_16_17.info()
```

```python
df_16_17 = df_16_17.round(2)
```

```python
df_16_17
```

```python
df_16_17.to_csv("2016_2017-building-energy-benchmarking.csv", index=False)
```
